```python
# get all climate combinations
# mf_low
# dry scenarios
real_All = [1270, 1956, 1987, 2770, 3449, 3515, 3574, 4211, 4373, 4937]
dT_All = [5]
dP_All = [80]
dCV_All = [1.2]
demand_All = ['MF_Low']
combinations_dry = list(itertools.product(real_All, dT_All, dP_All, dCV_All, demand_All))
#print(combinations_dry)

# wet scenarios
real_All = [1270, 1956, 1987, 2770, 3449, 3515, 3574, 4211, 4373, 4937]
dT_All = [0]
dP_All = [100]
dCV_All = [1.0]
demand_All = ['MF_Low']
combinations_wet = list(itertools.product(real_All, dT_All, dP_All, dCV_All, demand_All))
combinations_low = combinations_dry + combinations_wet

# mf_high
# dry scenarios
real_All = [1270, 1956, 1987, 2770, 3449, 3515, 3574, 4211, 4373, 4937]
dT_All = [5]
dP_All = [80]
dCV_All = [1.2]
demand_All = ['MF_High']
combinations_dry = list(itertools.product(real_All, dT_All, dP_All, dCV_All, demand_All))
#print(combinations_dry)

# wet scenarios
real_All = [1270, 1956, 1987, 2770, 3449, 3515, 3574, 4211, 4373, 4937]
dT_All = [0]
dP_All = [100]
dCV_All = [1.0]
demand_All = ['MF_High']
combinations_wet = list(itertools.product(real_All, dT_All, dP_All, dCV_All, demand_All))
combinations_high = combinations_dry + combinations_wet

print(combinations_low)
```

```python
# import and process data- monthly
filepath = '/Volumes/OneTouch/AffordPaper1_Results/SA-DemandDemographics/' # change to location where SA results are saved
rof = 0.65
inf_order = ['desal', 'dpr', 'mcasr', 'transfer_soquel', 'transfer_sv']

# SA information
# list of options to loop over
PriceElasticity = ['High', 'Low'] # -0.5, 0
MultiFamily = ['High', 'Low']
SF_Demands_Multiplier = ['High', 'Low'] # 1.15, 0.85
SF_Households = ['High', 'Low'] #23,507 and 19,233
SF_Income = ['High', 'Low'] # 36% HI, 28% LI
list_dfs = []
for pe, mf, dem, house, income in itertools.product(PriceElasticity, MultiFamily, SF_Demands_Multiplier, SF_Households, SF_Income):

    print('price elasticity: {}, multi-family demands: {}, single-family demands: {}, single-family households: {}, SF income: {}'.format(pe, mf, dem, house, income))
    
    # get combinations to use
    if mf == 'High':
        combinations_all = combinations_high
    elif mf == 'Low':
        combinations_all = combinations_low
    else:
        print('error')
        
    # define name to add to filenames
    name_add = f"PE={pe}_MF={mf}_Dem={dem}_House={house}_Income={income}_"
    # get monthly dataframe
    df_monthly = pf.aggregate_sows_for_dd_monthly(filepath, rof, inf_order, combinations_all, name_add)
    list_dfs.append(df_monthly)

df_all = pd.concat(list_dfs, ignore_index=False)
df_all.to_csv(filepath + 'df_monthly_SA_DD.csv', index=False)
```
